```python
# Drop the mostly-empty duplicate columns
df = df.drop(columns=['Title', 'alinknormal_URL', 'aiconalt', 'asizesmall', 'Price'])

# Rename the populated columns to clean, meaningful names
df = df.rename(columns={
    'Title1': 'Title',
    'alinknormal_URL1': 'Reviews_Link',
    'aiconalt1': 'Stars',
    'asizesmall1': 'Review_Count',
    'Price1': 'Price',
    'zgbdgtext': 'Rank'
})

df.columns.tolist()
```

```python
df.columns.tolist()
```

```python
#Check the null

df[df['Title'].isnull()]
```

```python
#drop the row

df = df.dropna(subset=['Title'])
df.shape
```

```python
#Remove duplicated rows

df.duplicated().sum()
df = df.drop_duplicates()
df.shape
```

```python
#convert stars to number

df['Stars'] = df['Stars'].str.extract(r'(\d+\.?\d*)').astype(float)
df['Stars'].head()
```

```python
#convert price to numbers

df['Price'] = df['Price'].str.replace('NGN', '', regex=False).str.replace(',', '', regex=False).str.strip().astype(float)
df['Price'].head()
```

```python
# BUSINESS QUESTIONS
# 1. What does the price distribution look like across Best Sellers? (histogram)
# 2. Is there a visible relationship between price and popularity (review count)? (scatterplot)
# 3. Which subcategory (if tracked) has the widest price range?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.regplot(data=df, x='Price', y='Review_Count', scatter_kws={'alpha':0.6})
plt.title('Price vs. Review Count — Amazon Home & Kitchen Best Sellers')
plt.xlabel('Price (NGN)')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=15, kde=True)
plt.title('Distribution of Prices — Amazon Home & Kitchen Best Sellers')
plt.xlabel('Price (NGN)')
plt.show()
```

```python
# 2. Is there a visible relationship between price and popularity (review count)? (scatterplot)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Price', y='Review_Count')
plt.title('Price vs. Review Count — Amazon Home & Kitchen Best Sellers')
plt.xlabel('Price (NGN)')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
#Since Stars had such a narrow range (4.2–4.8), 
# a boxplot showing that tight clustering (with any outlier points) 

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Stars'])
plt.title('Distribution of Star Ratings — Amazon Best Sellers')
plt.xlabel('Stars')
plt.show()
```
